```python
import numpy as np
import pandas as pd

from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv('housing_price_cleaned.csv')
```

```python
df.head()
```

```python
training_data = df.iloc[:1460, :]
testing_data = df.iloc[1460:]
```

```python
testing_data.head()
```

```python
Xtrain = training_data.drop('SalePrice', axis = 1)
Xtrain.shape
```

```python
Ytrain = training_data['SalePrice']
Ytrain.shape
```

```python
Xtest = testing_data.drop('SalePrice', axis = 1)
Xtest.shape
```

```python
#by using Random Forest Regressor
rfr = RandomForestRegressor(criterion = 'squared_error')
grid1 = {'max_depth':[5, 6, 7], 'n_estimators':[10, 50, 100]}

bestRFR = GridSearchCV(rfr, grid1, cv=5)
bestRFR.fit(Xtrain, Ytrain)
bestRFR.best_estimator_
```

```python
bestRFR.score(Xtrain, Ytrain)
```

```python
Ypred = bestRFR.predict(Xtest)
Ypred
```
